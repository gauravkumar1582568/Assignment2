```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3,10,100)

# first line
y = (((3/4)*x)+(-1))
plt.plot(x, y, '-r', label='3X-4Y=4')

#second line
y = (((21/123)*x)+0)
plt.plot(x, y, '-g', label='21X-123Y=0')

#third line
y = (((3/4)*x)+(-19/4))
plt.plot(x, y, '-b', label='3X-4Y-19=0')


#Fourth line
y = (((21/123)*x)+(-228/123))
plt.plot(x, y, '-y', label='21X-123Y-228=0')

plt.title('Graph of parallelogram created by four line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
